```python
df_train = pd.read_csv('https://stepik.org/media/attachments/course/4852/train_iris.csv', index_col=0)
```

```python
df_test = pd.read_csv('https://stepik.org/media/attachments/course/4852/test_iris.csv', index_col=0)
```

```python
x_train = df_train.drop(columns='species')
y_train = df_train.species

x_test = df_test.drop(columns='species')
y_test = df_test.species
```

```python
scores_data = pd.DataFrame()
```

```python
max_depth_values = range(1,100)
np.random.seed(0)
```

```python
for max_depth in max_depth_values:
    clf = tree.DecisionTreeClassifier(max_depth=max_depth)
    clf.fit(x_train, y_train)
    train_score = clf.score(x_train, y_train)
    y_pred = clf.predict(x_test)
    test_score = accuracy_score(y_test, y_pred)
    cv_score = cross_val_score(clf, x_train,y_train, cv=5).mean()

    scores_temp = pd.DataFrame({'max_depth':[max_depth],'train_score':[train_score],'test_score':[test_score],'cv_score':[cv_score]})

    scores_data = pd.concat([scores_data, scores_temp])
```

```python
scores_data = pd.melt(scores_data, id_vars=['max_depth'], value_vars=['train_score','test_score','cv_score'], var_name='set_type',
                      value_name='score')
```

```python
import seaborn as sns
```

```python
sns.lineplot(scores_data, x='max_depth', y='score', hue='set_type')
```

```python
import pandas as pd
import numpy as np
import patsy
import sklearn
import statsmodels.api as sm
from statsmodels.genmod.generalized_linear_model import GLM
from statsmodels.genmod import families
```

```python
df = pd.read_csv('Documents/Datasets/covid 19/covid.csv')
```

```python
df.head()
```

```python
df = df.rename(columns={'Region of residence':'region',
                        'Age of Subject':'age',
                        'Medium for online class':'device',
                        'Rating of Online Class experience':'rating',
                        'Time spent on self study':'time_selfstudy' ,
                       })
```

```python
formula = 'rating ~ C(time_selfstudy) + C(device) + C(time_selfstudy):C(device)'
```

```python
x, y = patsy.dmatrices(formula, data=df, return_type='dataframe')
```

```python
model = sm.GLM(y, x, family=families.Multinomial()).fit()
```

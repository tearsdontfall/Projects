```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('~/Old_Projects/Alerting Bot/spend_data.csv')
```

```python
df = df[df.offer.isna() == False]
```

```python
# df = df[df.offer.str.contains('yaeda_restaurant|dodo|delivery|hoff', case=False) == True]
```

```python
def metrics(df: pd.DataFrame) -> pd.DataFrame :
    df['CPM'] = (df.spend_usd / df.impressions * 1000).round(2)
    df['CPI'] = (df.spend_usd / df.installs).round(2)
    df['CPC'] = (df.spend_usd / df.clicks).round(2)
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    df.fillna(0, inplace=True)
    return df
```

```python
df = metrics(df)
```

```python
#оставляем только не нулевые значения
df = df[(df.CPM + df.CPI + df.CPI) > 0]
```

```python
df = df[df.business_dt < '2024-09-26']
```

```python
date = '2024-09-25'
```

```python
df['is_target'] = np.where(df.business_dt == '2024-09-25', 1,0)
```

```python
values = pd.DataFrame()
```

```python
for offer in df.offer.unique():
    for network in df.network.unique():
        if df[(df.network == network) & (df.offer == offer)
                        & (df[(df.network == network) & (df.offer == offer)].CPM > df[(df.network == network) & (df.offer == offer)].CPM.quantile(0.95))
                ].is_target.sum() > 0:
                    # print(f'extremum!!!! {offer}-{network}')
                    temp = pd.DataFrame({'offer':offer,'network':network, 
                                        'CPM':df[(df.network == network) & (df.offer == offer) & (df.business_dt ==date)].CPM.mean()}, index=[0])
                    values = pd.concat([values,temp])
        else:
            print(f'no extremums found for {offer}-{network}')
```
